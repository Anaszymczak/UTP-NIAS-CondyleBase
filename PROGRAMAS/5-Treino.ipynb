InferÃªncias usando o prÃ³prio dataset

```python
#InferÃªncia no dataset

import random
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np

# DiretÃ³rios das ROIs (ajuste se o nome for outro)
DIR_A_ROI = BASE_DIR / 'A_ROI'  # suposto Achatamento
DIR_N_ROI = BASE_DIR / 'N_ROI'  # suposto Normal

VALID_EXTS = {'.png', '.jpg', '.jpeg', '.bmp', '.tif', '.tiff'}

imagens_a = [p for p in DIR_A_ROI.rglob('*') if p.suffix.lower() in VALID_EXTS]
imagens_n = [p for p in DIR_N_ROI.rglob('*') if p.suffix.lower() in VALID_EXTS]

print(f"ğŸ“‚ A_ROI (esperado Achatamento): {len(imagens_a)} imagens")
print(f"ğŸ“‚ N_ROI (esperado Normal):      {len(imagens_n)} imagens")

# Quantos exemplos sortear de cada pasta
NUM_A = min(50, len(imagens_a))
NUM_N = min(50, len(imagens_n))

amostras_a = random.sample(imagens_a, NUM_A) if NUM_A > 0 else []
amostras_n = random.sample(imagens_n, NUM_N) if NUM_N > 0 else []

print(f"ğŸ¯ Amostras A_ROI: {NUM_A}")
print(f"ğŸ¯ Amostras N_ROI: {NUM_N}")

# ---------------------------------------------------------
# FunÃ§Ã£o auxiliar: roda prediÃ§Ã£o e retorna melhor classe
# conf_th  = limiar para o YOLO gerar boxes
# conf_min = limiar REAL que quer usar
# ---------------------------------------------------------
def predicao_principal(img_path, conf_th=0.1, conf_min=0.5):
    results = model_v2.predict(
        source=str(img_path),
        imgsz=640,
        conf=conf_th,
        verbose=False
    )
    r = results[0]

    # Se nÃ£o tem nenhuma box, jÃ¡ retorna como sem detecÃ§Ã£o
    if r.boxes is None or len(r.boxes) == 0:
        return None, None, None, r

    # Pega TODAS as confidÃªncias
    confs = r.boxes.conf.cpu().numpy()

    # Ãndices das boxes que passaram do limiar REAL
    valid_idxs = np.where(confs >= conf_min)[0]

    # Se nenhuma box passou do limiar:
    # - zera as boxes do r (pra nÃ£o desenhar nada)
    # - retorna cls_name = None
    if len(valid_idxs) == 0:
        r.boxes = r.boxes[valid_idxs]  # isso gera um boxes "vazio"
        return None, None, None, r

    # Dentro das vÃ¡lidas, pega a mais confiante
    best_local_idx = int(np.argmax(confs[valid_idxs]))
    idx = int(valid_idxs[best_local_idx])

    # MantÃ©m APENAS essa box em r.boxes,
    # assim o r.plot() sÃ³ desenha ela
    r.boxes = r.boxes[idx:idx+1]

    box = r.boxes[0]
    cls_id = int(box.cls)
    cls_name = model_v2.names[cls_id]
    score = float(box.conf)

    return cls_id, cls_name, score, r


# Montar lista: (img_path, gt_label_nome)
amostras_rotuladas = []
for p in amostras_a:
    amostras_rotuladas.append((p, 'Achatado'))
for p in amostras_n:
    amostras_rotuladas.append((p, 'Normal'))

num = len(amostras_rotuladas)
cols = 4
rows = (num + cols - 1) // cols

fig = plt.figure(figsize=(4 * cols, 4 * rows))

acertos = 0
erros = 0
sem_det = 0

for i, (img_path, gt_nome) in enumerate(amostras_rotuladas, start=1):

    #  ontrola o limiar REAL (0.5)
    cls_id, cls_name, score, r = predicao_principal(img_path, conf_th=0.1, conf_min=0.5)

    # Imagem com a box desenhada (se tiver)
    img_bgr = r.plot()
    img_rgb = img_bgr[..., ::-1]

    #print(gt_nome, " - ", cls_name)
    if cls_name is None:
        status = "Sem detecÃ§Ã£o (conf < 0.50 ou nenhuma box)"
        sem_det += 1
        titulo = f"{img_path.name}\nGT={gt_nome} | {status}"
    else:
        if cls_name == gt_nome:
            status = "âœ… ACERTO"
            acertos += 1
        else:
            status = "âŒ ERRO"
            erros += 1

        titulo = f"{img_path.name}\nGT={gt_nome} | Pred={cls_name} ({score:.2f}) | {status}"

    plt.subplot(rows, cols, i)
    plt.imshow(img_rgb)
    plt.title(titulo, fontsize=8)
    plt.axis('off')

plt.tight_layout()
plt.show()

print("========= RESUMO =========")
print(f"âœ… Acertos:       {acertos}")
print(f"âŒ Erros:         {erros}")
print(f"âšª Sem detecÃ§Ã£o:  {sem_det}")
print(f"Total exemplos:  {num}")
```

 Experimentos repetidos (DATASET) (100 rodadas com 50A + 50N)

```python
import random
import numpy as np
from pathlib import Path


# CONFIGURAÃ‡ÃƒO

DIR_A = BASE_DIR / 'A_ROI'
DIR_N = BASE_DIR / 'N_ROI'

VALID_EXTS = {'.png', '.jpg', '.jpeg', '.bmp', '.tif', '.tiff'}

# Carrega todas as imagens internas
imgs_A = [p for p in DIR_A.rglob('*') if p.suffix.lower() in VALID_EXTS]
imgs_N = [p for p in DIR_N.rglob('*') if p.suffix.lower() in VALID_EXTS]

print(f"Total A: {len(imgs_A)} | Total N: {len(imgs_N)}")


# FunÃ§Ã£o de prediÃ§Ã£o

def predict_one(img_path, conf_min=0.5):

    results = model_v2.predict(
        source=str(img_path),
        imgsz=640,
        conf=0.1,
        verbose=False
    )
    r = results[0]

    if r.boxes is None or len(r.boxes) == 0:
        return None

    confs = r.boxes.conf.cpu().numpy()
    valid_idxs = np.where(confs >= conf_min)[0]

    if len(valid_idxs) == 0:
        return None

    best_local = valid_idxs[np.argmax(confs[valid_idxs])]
    cls_id = int(r.boxes[best_local].cls)
    return model_v2.names[cls_id]  # "Achatado" ou "Normal"


# FunÃ§Ã£o de execuÃ§Ã£o de 1 experimento

def rodar_experimento():

    # sorteia 50A + 50N
    sa = random.sample(imgs_A, min(50, len(imgs_A)))
    sn = random.sample(imgs_N, min(50, len(imgs_N)))

    acertos = 0
    erros = 0
    sem_det = 0

    # Avaliar A (GT = Achatado)
    for p in sa:
        pred = predict_one(p)
        if pred is None:
            sem_det += 1
        elif pred == "Achatado":
            acertos += 1
        else:
            erros += 1

    # Avaliar N (GT = Normal)
    for p in sn:
        pred = predict_one(p)
        if pred is None:
            sem_det += 1
        elif pred == "Normal":
            acertos += 1
        else:
            erros += 1

    return acertos, erros, sem_det

# Rodar 100 experimentos
N_EXP = 100

lista_acertos = []
lista_erros = []
lista_sem_det = []
lista_acuracia = []

for i in range(N_EXP):
    a, e, sd = rodar_experimento()
    total_validas = a + e + sd
    acc = a / total_validas if total_validas > 0 else 0

    lista_acertos.append(a)
    lista_erros.append(e)
    lista_sem_det.append(sd)
    lista_acuracia.append(acc)

    print(f"Rodada {i+1}: Acertos={a}, Erros={e}, SemDet={sd}, AcurÃ¡cia={acc*100:.2f}%")
```

```python
print("\n========== RESULTADO FINAL ==========")
print(f"MÃ©dia Acertos:       {np.mean(lista_acertos):.2f} Â± {np.std(lista_acertos):.2f}")
print(f"MÃ©dia Erros:         {np.mean(lista_erros):.2f} Â± {np.std(lista_erros):.2f}")
print(f"MÃ©dia Sem DetecÃ§Ã£o:  {np.mean(lista_sem_det):.2f} Â± {np.std(lista_sem_det):.2f}")
print(f"AcurÃ¡cia MÃ©dia:      {np.mean(lista_acuracia)*100:.2f}% Â± {np.std(lista_acuracia)*100:.2f}%")
```
