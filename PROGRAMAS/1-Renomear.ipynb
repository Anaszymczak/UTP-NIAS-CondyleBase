```python
# Montar Drive e imports
from google.colab import drive
drive.mount('/content/drive')

import cv2, numpy as np
from pathlib import Path
```

```python
# Config diret√≥rios
BASE_DIR = Path('/content/drive/MyDrive/UTP_ATM_V1')
DIR_A_ORIGINAL = BASE_DIR / 'A_ORIGINAL'
DIR_N_ORIGINAL = BASE_DIR / 'N_ORIGINAL'
VALID_EXTS = {'.png','.jpg','.jpeg','.bmp','.tif','.tiff'}
DRY_RUN = False

LOG_FILE = BASE_DIR / 'rastreio_processamento.txt'
def append_log(line: str):
    with open(LOG_FILE, 'a', encoding='utf-8') as f:
        f.write((("[SIMULACAO] " if DRY_RUN else "") + line).rstrip() + '\n')

def rename_originals_to_imgXN(folder: Path, suffix: str):
    imgs = [p for p in folder.rglob('*') if p.suffix.lower() in VALID_EXTS]
    imgs.sort()
    # tempor√°rio para evitar colis√£o
    tmps = []
    for i, p in enumerate(imgs):
        tmp = p.with_name(f'__REN_TMP__{i:06d}{p.suffix.lower()}')
        if not DRY_RUN: p.rename(tmp)
        tmps.append((p, tmp))
    # nomes finais
    for i, (orig, tmp) in enumerate(tmps, start=1):
        new_name = f'img{i}{suffix}{tmp.suffix.lower()}'
        new_path = tmp.with_name(new_name)
        if not DRY_RUN: tmp.rename(new_path)
        append_log(f"{Path(new_name).stem} <- renomeada de {orig.name}")
    print(f"‚úÖ {folder.name}: {len(imgs)} arquivo(s) renomeado(s).")
```

```python
# Execu√ß√£o
assert DIR_A_ORIGINAL.exists() and DIR_N_ORIGINAL.exists(), "Pastas A_ORIGINAL/N_ORIGINAL n√£o encontradas."
rename_originals_to_imgXN(DIR_A_ORIGINAL, 'A')
rename_originals_to_imgXN(DIR_N_ORIGINAL, 'N')
print("üìÑ Rastreio:", LOG_FILE)
```
